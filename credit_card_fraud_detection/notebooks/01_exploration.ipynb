```python
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

**Observations:**

* Most features (`V1`â€“`V28`) have very **low correlation** with each other â€” the heatmap is mostly light colors.
* A few features (like `V17`, `V14`, `V12`, `V10`) show **moderate correlation** (positive or negative) with `Class` â€” visible as slightly stronger blue or red cells in the `Class` row/column.
* `Time` and `Amount` have **very weak correlation** with `Class`.
* Self-correlations dominate the diagonal (always 1.0), which makes the rest of the variations harder to see.

**Conclusions:**

* The fraud signal is **spread across multiple PCA-transformed features**, none of which alone has a strong correlation â€” meaning classification will require **multivariate analysis** (like Random Forest, XGBoost) rather than relying on one feature.
* `V17`, `V14`, `V12`, and `V10` are worth watching â€” they might contribute more to the modelâ€™s predictive power.
* Correlation-based feature selection may not be very effective here; models that can capture **non-linear relationships** are better suited.
